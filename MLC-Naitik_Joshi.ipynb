## 2.3 Encoding

I am using 2 types of encoding to convert string data type into the numeric data type.
1. One-Hot Encoding
2. Target Encoding

I have columns with numerous unique values, such as standard make and model, which is why I am using the target encoding method. Because of these two columns, there will be too many columns if I use a one hot encoding for them.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ohe_cat_cols = ['standard_colour','body_type','fuel_type','vehicle_condition']
```

```python
# Categorical Encoding with One-Hot Encoding
one_hot_encoded_df = pd.get_dummies(adv[ohe_cat_cols])
```

```python
one_hot_encoded_df.head()
```

```python
# Categorical Encoding with Target Encoding
for col in ['standard_make', 'standard_model']:
    target_means = adv.groupby(col)['price'].mean()
    adv[col + '_encoded'] = adv[col].map(target_means)
```

```python
# Merging both encoded dataset
data_encoded = pd.concat([adv.drop(ohe_cat_cols + ['standard_make', 'standard_model'], axis=1), 
                          one_hot_encoded_df], axis=1)
```

```python
data_encoded.head()
```
